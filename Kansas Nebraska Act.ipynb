```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    'Party': ['Democrats', 'Whigs', 'Free Soilers', 'Independent', 'Independent Democrat'],
    'Number of Representatives': [157, 71, 4, 1, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set Seaborn style
sns.set(style="whitegrid")

# Bar chart
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Party', y='Number of Representatives', data=df, palette="viridis")
plt.title('Number of Representatives by Political Party', fontsize=16)
plt.xlabel('Political Party', fontsize=14)
plt.ylabel('Number of Representatives', fontsize=14)

# Customize bar labels with whole numbers
for p in bar_plot.patches:
    bar_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12, color='black')

plt.show()

# Pie chart
plt.figure(figsize=(8, 8))

# Adjust autopct to include labels and avoid overlapping percentages
plt.pie(df['Number of Representatives'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))

plt.title('Distribution of Representatives by Political Party', fontsize=16)

# Add legend to the pie chart
plt.legend(title='Party')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
parties = ['Northern Democrats', 'Northern Whigs', 'Southern Democrats', 'Southern Whigs']
votes_in_favor = [44, 0, 57, 12]
votes_against = [42, 45, 2, 7]

# Create a bar graph
bar_width = 0.35
index = range(len(parties))

fig, ax = plt.subplots()
bar1 = ax.bar(index, votes_in_favor, bar_width, label='In Favor')
bar2 = ax.bar([i + bar_width for i in index], votes_against, bar_width, label='Against')

# Labeling and formatting
ax.set_xlabel('Parties')
ax.set_ylabel('Number of Votes')
ax.set_title('Final House Vote on the Bill')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(parties, rotation=45, ha="right")  # Rotate labels for better readability
ax.legend()

# Display the graph
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# Data
years = [1790, 1800, 1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900]
population = [0, 0, 0, 0, 0, 0, 0, 107206, 364399, 996096, 1428108, 1470495]

# Create the graph
plt.figure(figsize=(10, 6))

# Plot the line graph
plt.plot(years, population, marker='o', color='b', linestyle='-', linewidth=2, label='Population')

# Fill the area under the line
plt.fill_between(years, population, color='skyblue', alpha=0.4)

# Add a note about missing data before 1860s
plt.text(1825, max(population) * 0.9, 'Population data not recorded before 1860s\nas Kansas was not yet a state',
         fontsize=12, color='gray', ha='center')

# Labeling and formatting
plt.title('Population Growth of Kansas (1790-1900)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)

# Format y-axis labels without scientific notation
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.ticklabel_format(axis='y', style='plain')

plt.legend()  # Display legend
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from datetime import date, timedelta
import numpy as np

# Events and their corresponding dates
events = {
    "Missouri Compromise": date(1820, 3, 6),
    "Wilmot Proviso": date(1846, 8, 8),
    "Compromise of 1850": date(1850, 9, 20),
    "Uncle Tom's Cabin Published": date(1852, 3, 20),
    "Kansas-Nebraska Act": date(1854, 5, 30)
}

min_date = min(events.values()) - timedelta(days=365)
max_date = max(events.values()) + timedelta(days=365)

labels = list(events.keys())
# labels with associated dates
labels = ['{0:%d %b %Y}:\n{1}'.format(d, l) for l, d in zip(labels, events.values())]

fig, ax = plt.subplots(figsize=(8, 12), constrained_layout=True)
_ = ax.set_xlim(-3, 3)
_ = ax.set_ylim(min_date, max_date)
_ = ax.axvline(0, ymin=0.05, ymax=0.95, c='deeppink', zorder=1)

_ = ax.scatter(np.zeros(len(events)), list(events.values()), s=120, c='palevioletred', zorder=2)
_ = ax.scatter(np.zeros(len(events)), list(events.values()), s=30, c='darkmagenta', zorder=3)

label_offsets = np.linspace(2.0, -2.0, len(events))
for i, (l, d) in enumerate(zip(labels, events.values())):
    d = d - timedelta(days=90)
    align = 'right' if i % 2 == 0 else 'left'
    _ = ax.text(label_offsets[i], d, l, ha=align, fontfamily='serif', fontweight='bold', color='royalblue', fontsize=10)

stems = np.linspace(2.0, -2.0, len(events))
_ = ax.hlines(list(events.values()), 0, stems, color='darkmagenta')

# hide lines around chart
for spine in ["left", "top", "right", "bottom"]:
    _ = ax.spines[spine].set_visible(False)

# hide tick labels
_ = ax.set_xticks([])
_ = ax.set_yticks([])

_ = ax.set_title('Events Leading to the Kansas-Nebraska Act', fontweight="bold", fontfamily='serif', fontsize=16,
                 color='royalblue')

plt.show()
```
